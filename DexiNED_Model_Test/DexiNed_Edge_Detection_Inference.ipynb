```python
# ============================================
# CELL 10: DEFINE PREPROCESSING FUNCTIONS
# Image preprocessing utilities for DexiNed
# ============================================

print("\n" + "="*60)
print("DEFINING PREPROCESSING FUNCTIONS")
print("="*60)

def preprocess_image(image_path, target_height=512, target_width=512):
    """
    Load and preprocess image for DexiNed inference.
    
    CRITICAL: Dimensions MUST be divisible by 16!
    
    Args:
        image_path: Path to input image
        target_height: Target height (must be divisible by 16)
        target_width: Target width (must be divisible by 16)
    
    Returns:
        tuple: (preprocessed_tensor, original_rgb_image, new_shape)
    """
    # Load image (OpenCV loads as BGR)
    img = cv2.imread(str(image_path), cv2.IMREAD_COLOR)
    if img is None:
        raise ValueError(f"Could not load image: {image_path}")
    
    # Store original for visualization
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    original_shape = img.shape[:2]
    
    # Ensure dimensions are divisible by 16
    target_height = (target_height // 16) * 16
    target_width = (target_width // 16) * 16
    
    # Resize image
    img_resized = cv2.resize(img, (target_width, target_height))
    
    # Convert BGR to RGB (DexiNed expects RGB)
    img_rgb_resized = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
    
    # Normalize: convert to float and scale to [0, 1]
    img_normalized = img_rgb_resized.astype(np.float32) / 255.0
    
    # Convert to PyTorch tensor: (H, W, C) -> (C, H, W)
    img_tensor = torch.from_numpy(img_normalized).permute(2, 0, 1)
    
    # Add batch dimension: (C, H, W) -> (1, C, H, W)
    img_tensor = img_tensor.unsqueeze(0)
    
    return img_tensor, img_rgb, (target_height, target_width)

def run_inference(model, img_tensor, device):
    """
    Run DexiNed inference on preprocessed image.
    
    Args:
        model: DexiNed model
        img_tensor: Preprocessed image tensor
        device: torch device
    
    Returns:
        List of 7 edge maps (6 side outputs + 1 fused)
    """
    img_tensor = img_tensor.to(device)
    
    with torch.no_grad():
        preds = model(img_tensor)
    
    return preds

def extract_fused_edge(predictions, threshold=None):
    """
    Extract the fused edge map (best quality).
    
    Args:
        predictions: Model output list
        threshold: Optional binary threshold (e.g., 0.5)
    
    Returns:
        numpy array (H, W) with values in [0, 1] or binary
    """
    fused = predictions[-1].squeeze().cpu().numpy()
    
    if threshold is not None:
        fused = (fused > threshold).astype(np.float32)
    
    return fused

print("\n✓ Preprocessing functions defined:")
print("  - preprocess_image()")
print("  - run_inference()")
print("  - extract_fused_edge()")
print("="*60)
```

```python
# ============================================
# CELL 11: DEFINE VISUALIZATION FUNCTIONS
# Utilities for displaying edge detection results
# ============================================

print("\n" + "="*60)
print("DEFINING VISUALIZATION FUNCTIONS")
print("="*60)

def visualize_comparison(original_img, edge_map, figsize=(16, 8), auto_normalize=True):
    """
    Display side-by-side comparison of original and edge map.
    
    Args:
        auto_normalize: If True, normalizes edge map to [0, 1] for better visibility
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    ax1.imshow(original_img)
    ax1.set_title('Original Image', fontsize=14, fontweight='bold')
    ax1.axis('off')
    
    # Auto-normalize if values are too small
    if auto_normalize and edge_map.max() > 0:
        edge_display = (edge_map - edge_map.min()) / (edge_map.max() - edge_map.min())
        title = f'Detected Edges (range: {edge_map.min():.3f}-{edge_map.max():.3f})'
    else:
        edge_display = edge_map
        title = 'Detected Edges'
    
    ax2.imshow(edge_display, cmap='gray', vmin=0, vmax=1)
    ax2.set_title(title, fontsize=14, fontweight='bold')
    ax2.axis('off')
    
    plt.tight_layout()
    plt.show()

def visualize_all_outputs(original_img, predictions, auto_normalize=True):
    """
    Display all 7 DexiNed outputs in a grid.
    Shows 6 side outputs + 1 fused output.
    
    Args:
        auto_normalize: If True, normalizes each edge map independently
    """
    edge_maps = [pred.squeeze().cpu().numpy() for pred in predictions]
    
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    axes = axes.flatten()
    
    # Original image
    axes[0].imshow(original_img)
    axes[0].set_title('Original Image', fontsize=12, fontweight='bold')
    axes[0].axis('off')
    
    # 6 side outputs
    for i in range(6):
        edge = edge_maps[i]
        # Normalize each output independently
        if auto_normalize and edge.max() > 0:
            edge = (edge - edge.min()) / (edge.max() - edge.min())
        axes[i+1].imshow(edge, cmap='gray', vmin=0, vmax=1)
        axes[i+1].set_title(f'Side Output {i+1}', fontsize=11)
        axes[i+1].axis('off')
    
    # Fused output (best quality)
    edge = edge_maps[6]
    if auto_normalize and edge.max() > 0:
        edge = (edge - edge.min()) / (edge.max() - edge.min())
    axes[7].imshow(edge, cmap='gray', vmin=0, vmax=1)
    axes[7].set_title('Fused Output (Best)', fontsize=12, fontweight='bold')
    axes[7].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return edge_maps

def create_edge_overlay(original_img, edge_map, edge_color=(0, 255, 0)):
    """
    Overlay edges on original image with specified color.
    
    Args:
        original_img: RGB image (H, W, 3)
        edge_map: Edge map (H, W) with values [0, 1]
        edge_color: RGB tuple for edge color
    """
    # Ensure same dimensions
    if edge_map.shape[:2] != original_img.shape[:2]:
        edge_map = cv2.resize(edge_map, 
                             (original_img.shape[1], original_img.shape[0]))
    
    # Create RGB overlay
    overlay = original_img.copy().astype(np.float32)
    
    # Normalize edge map to [0, 1]
    edge_normalized = edge_map / edge_map.max() if edge_map.max() > 0 else edge_map
    
    # Apply colored overlay
    for i in range(3):
        overlay[:, :, i] = overlay[:, :, i] * (1 - edge_normalized) + \
                          edge_color[i] * edge_normalized
    
    overlay = np.clip(overlay, 0, 255).astype(np.uint8)
    return overlay

print("\n✓ Visualization functions defined (with auto-normalization):")
print("  - visualize_comparison()")
print("  - visualize_all_outputs()")
print("  - create_edge_overlay()")
print("="*60)
```

```python
# ============================================
# CELL 16: BATCH PROCESSING FUNCTION
# Process multiple images with progress tracking
# ============================================

print("\n" + "="*60)
print("DEFINING BATCH PROCESSING FUNCTION")
print("="*60)

def batch_process_images(
    input_dir='data',
    output_dir='output',
    target_size=512,
    save_visualizations=True
):
    """
    Process all images in a directory.
    
    Args:
        input_dir: Directory containing input images
        output_dir: Directory for output edge maps
        target_size: Processing resolution (must be divisible by 16)
        save_visualizations: Save side-by-side comparisons
    
    Returns:
        dict: Processing statistics
    """
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)
    
    # Find all images
    extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp']
    image_paths = []
    for ext in extensions:
        image_paths.extend(input_dir.glob(ext))
        image_paths.extend(input_dir.glob(ext.upper()))
    
    print(f"\nFound {len(image_paths)} images to process")
    
    stats = {'total': len(image_paths), 'successful': 0, 'failed': 0}
    
    # Process each image
    for img_path in tqdm(image_paths, desc="Processing images"):
        try:
            # Preprocess
            img_tensor, img_original, _ = preprocess_image(
                img_path, target_size, target_size
            )
            
            # Inference
            preds = run_inference(model, img_tensor, device)
            edge_map = extract_fused_edge(preds)
            
            # Save edge map
            edge_path = output_dir / f"{img_path.stem}_edges.png"
            edge_uint8 = (edge_map * 255).astype(np.uint8)
            cv2.imwrite(str(edge_path), edge_uint8)
            
            # Save visualization
            if save_visualizations:
                vis_path = output_dir / f"{img_path.stem}_comparison.png"
                fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
                ax1.imshow(img_original)
                ax1.set_title('Original')
                ax1.axis('off')
                ax2.imshow(edge_map, cmap='gray')
                ax2.set_title('Edges')
                ax2.axis('off')
                plt.tight_layout()
                plt.savefig(vis_path, dpi=150, bbox_inches='tight')
                plt.close(fig)
            
            # Cleanup
            del img_tensor, preds, edge_map
            stats['successful'] += 1
            
        except Exception as e:
            print(f"\n❌ Failed on {img_path.name}: {str(e)}")
            stats['failed'] += 1
    
    # Cleanup GPU memory
    torch.cuda.empty_cache()
    
    return stats

print("\n✓ Batch processing function defined")
print("\nUsage:")
print("  stats = batch_process_images(")
print("      input_dir='data',")
print("      output_dir='output',")
print("      target_size=512")
print("  )")
print("="*60)
```

```python
# ============================================
# CELL 18: GPU MEMORY MANAGEMENT
# Monitor and clear GPU memory
# ============================================

print("\n" + "="*60)
print("GPU MEMORY MANAGEMENT")
print("="*60)

def check_gpu_memory():
    """Display current GPU memory usage"""
    if not torch.cuda.is_available():
        print("\nCUDA not available")
        return
    
    torch.cuda.empty_cache()
    allocated = torch.cuda.memory_allocated(0) / 1e9
    reserved = torch.cuda.memory_reserved(0) / 1e9
    total = torch.cuda.get_device_properties(0).total_memory / 1e9
    
    print(f"\nGPU Memory Status:")
    print(f"  Allocated: {allocated:.2f} GB")
    print(f"  Reserved: {reserved:.2f} GB")
    print(f"  Free: {(total - reserved):.2f} GB")
    print(f"  Total: {total:.2f} GB")
    print(f"  Usage: {(reserved/total*100):.1f}%")

def clear_gpu_memory():
    """Comprehensive GPU memory cleanup"""
    gc.collect()
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        torch.cuda.synchronize()
    print("\n✓ GPU memory cleared")

# Check current status
check_gpu_memory()

# Clear memory if needed
# clear_gpu_memory()

print("\n" + "="*60)
print("Memory management functions:")
print("  check_gpu_memory()  - Display current usage")
print("  clear_gpu_memory()  - Free cached memory")
print("="*60)
```
